```python
import hashlib

def calcular_hash(texto):
    """Función auxiliar para aplicar SHA256 a un string."""
    return hashlib.sha256(texto.encode('utf-8')).hexdigest()

def construir_arbol_merkle(transacciones):

    nivel_actual = [calcular_hash(tx) for tx in transacciones]

    print(f"Nivel Inicial (Hojas): {nivel_actual}")

    # Ciclo para subir "de abajo hacia arriba"
    # Continuamos mientras haya más de un hash en la lista
    while len(nivel_actual) > 1:
        nivel_superior = []

        # Procesamos en pares (de 2 en 2)
        for i in range(0, len(nivel_actual), 2):
            hoja_izquierda = nivel_actual[i]

            # Manejo de impares: si hay un número impar de elementos,
            # el último se duplica o se concatena consigo mismo.
            if i + 1 < len(nivel_actual):
                hoja_derecha = nivel_actual[i + 1]
            else:
                hoja_derecha = hoja_izquierda # Duplicar el último si es impar

            # Concatenamos y hasheamos (Hash0 + Hash1 -> Hash01)
            combinacion = hoja_izquierda + hoja_derecha
            nuevo_hash = calcular_hash(combinacion)

            nivel_superior.append(nuevo_hash)

            print(f"Combinando: {hoja_izquierda[:6]}... + {hoja_derecha[:6]}... -> {nuevo_hash[:6]}...")

        # El nivel superior se convierte en el nivel actual para la siguiente vuelta
        nivel_actual = nivel_superior
        print(f"Nivel intermedio completado: {len(nivel_actual)} nodos restantes.\n")

    # El último elemento restante es el Hash Raíz
    return nivel_actual[0]

# Se ejecuta

# Datos de ejemplo
mis_transacciones = ["hola", "Hola", "chao", "Chao"]

merkle_root = construir_arbol_merkle(mis_transacciones)

print("-" * 30)
print(f"MERKLE ROOT FINAL: {merkle_root}")
```
